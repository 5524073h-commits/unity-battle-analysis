```python

!pip install japanize-matplotlib

import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np


def to_mps(laps):
    """ラップタイム(200m)を秒速(m/s)に変換"""
    return [200 / l for l in laps]


jc_dist = np.arange(200, 2600, 200)

jc_k_laps = [12.7, 11.3, 12.1, 12.3, 12.1, 12.4, 12.4, 12.2, 12.1, 11.2, 11.3, 11.7]
jc_a_laps = [12.9, 10.8, 12.2, 12.3, 11.7, 11.6, 11.7, 11.6, 11.8, 11.4, 11.4, 11.2]
jc_e_laps = [13.2, 11.2, 11.8, 12.1, 12.0, 12.1, 12.1, 11.9, 11.5, 11.3, 11.3, 11.3]


as_k_laps = [13.2, 12.5, 12.7, 13.0, 13.2, 13.0, 12.5, 12.4, 13.1, 12.7]
as_a_laps = [12.8, 11.3, 11.5, 11.6, 11.8, 11.5, 11.6, 11.3, 11.5, 11.3]
as_e_laps = [12.4, 10.4, 11.5, 11.7, 11.7, 11.4, 11.4, 11.4, 11.6, 11.7]


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 14))

ax1.plot(jc_dist, to_mps(jc_k_laps), label='キタサンブラック (2016)', color='#e74c3c', marker='o', alpha=0.6)
ax1.plot(jc_dist, to_mps(jc_a_laps), label='アーモンドアイ (2018レコード)', color='#f1c40f', marker='p', linewidth=3)
ax1.plot(jc_dist, to_mps(jc_e_laps), label='イクイノックス (2023)', color='#2c3e50', marker='s', linewidth=4)
ax1.set_title('【2400m】ジャパンカップ 速度比較', fontsize=15, fontweight='bold')
ax1.set_ylabel('速度 (m/s)')
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend()

ax2.plot(as_dist, to_mps(as_k_laps), label='キタサンブラック (2017 不良)', color='#8e44ad', marker='o', alpha=0.6)
ax2.plot(as_dist, to_mps(as_a_laps), label='アーモンドアイ (2019)', color='#f1c40f', marker='p', linewidth=2)
ax2.plot(as_dist, to_mps(as_e_laps), label='イクイノックス (2023 レコード)', color='#2c3e50', marker='s', linewidth=4)
ax2.set_title('【2000m】天皇賞・秋 速度比較 (泥沼 vs レコード)', fontsize=15, fontweight='bold')
ax2.set_xlabel('走行距離 (m)')
ax2.set_ylabel('速度 (m/s)')
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.legend()

plt.tight_layout()
plt.show()
```
